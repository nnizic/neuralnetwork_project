<h3>Testiranje najboljeg modela na test podatcima</h3>
<h5>Test podatci su 3 klase po 250 slika</h5>

```python
from tensorflow.keras.models import load_model

# Load the saved model
model = load_model('iteracija_k5.h5')
model_tl = load_model('iteracija_l2.h5')
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Definiranje generatora za slike
test_datagen = ImageDataGenerator(rescale=1.0 / 255)

# Učitavanje test slka
test_generator = test_datagen.flow_from_directory(
    '../../Projekt_slike_128/TEST/Test/',  
    target_size=(128, 128),  
    batch_size=32,
    class_mode='categorical',
    shuffle=False  
)
```

```python
import numpy as np

# Predvidi klase za test slike
predictions = model.predict(test_generator)
predicted_classes = np.argmax(predictions, axis=1)
true_classes = test_generator.classes
class_labels = list(test_generator.class_indices.keys())
```

```python
from sklearn.metrics import accuracy_score

# Izračun točnosti za svaku klasu
for i, label in enumerate(class_labels):
    indices = np.where(true_classes == i)[0]
    class_accuracy = accuracy_score(true_classes[indices], predicted_classes[indices])
    print(f'Točnost za klasu {label}: {class_accuracy:.2f}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Compute the confusion matrix
cm = confusion_matrix(true_classes, predicted_classes)

# Plot the confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel('Predviđene')
plt.ylabel('Točne')
plt.title('Konfuzijska matrica')
plt.show()
```

```python
# Evaluiranje modela prema test podatcima
loss, accuracy = model.evaluate(test_generator)

# Ukupna točnost
print(f'Ukupna točnost modela: {accuracy:.2f}')
# Ukupni gubitak
print(f'Ukupni gubitak modela: {loss:.2f}')
```
